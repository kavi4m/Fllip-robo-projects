```python
import pandas as pd
from selenium import webdriver
import time
```

```python
browser = webdriver.Chrome('chromedriver.exe')
```

```python
def get_data(link,driver):
    driver.get(link)
    time.sleep(2)
    headline=driver.find_element_by_xpath('//h1[@class="artTitle font_faus"]').text.strip()
    #print(headline)
    
    vertical=link.split('/')[-4]
    #print(vertical)
    
    try:
        author1=driver.find_element_by_xpath('//div[@class="auth eventDone"]').text.strip()
    except:
        author1=''
    author2=driver.find_element_by_xpath('//span[@class="ag"]').text
    if author1!='':
        author=author1+','+author2
    else:
        author=author2
    #print(author)
    date=driver.find_element_by_xpath('//time').text.replace('Last Updated: ','')
    #print(date)
    description=driver.find_element_by_xpath('//div[@class="artText medium"]').text.strip()
    #print(description)
    return headline,date,vertical,author,description
```

```python
data={
    'date':[],
    'author':[],
    'vertical':[],
    'headline':[],
    'description':[]
    }


start=43405
for i in range(30):
    print(f'page:-{i}')
    try:
        browser.get(f"https://economictimes.indiatimes.com/archivelist/year-2018,month-11,starttime-{start}.cms")
        start+=1
        time.sleep(2)
        tds=browser.find_elements_by_xpath('//td[@class="contentbox5"]')[-2:]
        #print(len(tds))
        all_links=[]
        for td in tds:
            links=td.find_elements_by_xpath('.//li')
            links=[e.find_element_by_xpath('.//a').get_attribute('href') for e in links]
            all_links.extend(links)
        for link in all_links:
            try:
                #print(link)
                headline,date,vertical,author,description=get_data(link,browser)
                data['date'].append(date)
                data['headline'].append(headline)
                data['vertical'].append(vertical)
                data['author'].append(author)
                data['description'].append(description)
            except Exception as err1:
                print(err1)
                pass
        
    except Exception as err:
        print(err)
        pass
    #break
    
df=pd.DataFrame.from_dict(data)
df.to_excel('output_economic_times.xlsx',index=False)
```
